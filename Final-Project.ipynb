#### Plot a bar graph to show the distribution of different Credit Card Provider ('CC Provider') usage.

```python
# Needs more than one line of code (around 5 or 6 lines of code)
import matplotlib.pyplot as plt
import numpy as np
providers=ecom["CC Provider"].tolist()


arr=[]  
i=0
while i<(len(providers)):
    arr.append(providers[i])
    i=i+1



elements_count = {}
a={}
for element in arr:
    if element in elements_count:
        elements_count[element] += 1
    else:
        elements_count[element] = 1
        
a = dict(sorted(elements_count.items(), key=lambda item: item[1], reverse=True))
print(a)
x=[]
y=[]
for key, value in a.items():
   #print(f"{key}: {value}")
    x.append(key)
    y.append(value)

 


plt.bar(x=x,height=y,width=0.8)
plt.title('CC providers frequency')
plt.show()
```

##### Plot the above figures (bar graph and pie chart) in a one figure.

```python
# Needs more than one line of code (around 6 or 7 lines of code)
import matplotlib.pyplot as plt
import numpy as np
providers=ecom["CC Provider"].tolist()


arr=[]  
i=0
while i<(len(providers)):
    arr.append(providers[i])
    i=i+1



elements_count = {}
a={}
for element in arr:
    if element in elements_count:
        elements_count[element] += 1
    else:
        elements_count[element] = 1
        
a = dict(sorted(elements_count.items(), key=lambda item: item[1], reverse=True))

x=[]
y=[]
for key, value in a.items():
   #print(f"{key}: {value}")
    x.append(key)
    y.append(value)
plt.figure(figsize=[20,10])
plt.subplot(2,2,1)

plt.pie(x=y,labels=x,autopct='% 1.1f%%')
 

plt.subplot(2,2,2)

plt.bar(x=x,height=y,width=0.8)
plt.title('CC providers frequency')
plt.show()
```
